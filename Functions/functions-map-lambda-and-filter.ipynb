```python
#3. Check even
def is_even(n):
    return n % 2 == 0
```

```python
#4. Use filter with is_even
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(is_even, nums))
print(evens)
```

```python
#5. Lambda with filter
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
#6. Square function with map
def square(n):
    return n * n

squares = list(map(square, nums))
print(squares)
```

```python
#7. Lambda with map
squares = list(map(lambda x: x * x, nums))
print(squares)
```

```python
#8. Square only even numbers
result = list(map(lambda x: x * x, filter(lambda x: x % 2 == 0, nums)))
print(result)
```

```python
#9. Starts with vowel
def starts_with_vowel(s):
    return s[0].lower() in "aeiou"
```

```python
#10. Filter words starting with vowel
words = ["apple", "banana", "orange", "grape", "elephant"]
vowel_words = list(filter(starts_with_vowel, words))
print(vowel_words)
```

```python
#11. Lambda version
vowel_words = list(filter(lambda s: s[0].lower() in "aeiou", words))
print(vowel_words)
```

```python
#12. Length of string with map
def length(s):
    return len(s)

lengths = list(map(length, words))
print(lengths)
```

```python
#13. Lambda version
lengths = list(map(lambda s: len(s), words))
print(lengths)
```

```python
#14. Lengths of words starting with vowels
result = list(map(lambda s: len(s), filter(lambda s: s[0].lower() in "aeiou", words)))
print(result)
```

```python
#15. Squares of only odd numbers
def odd_squares(nums):
    return [x * x for x in nums if x % 2 != 0]
```

```python
#16. Reusable map function
def apply_func(lst, func):
    return list(map(func, lst))
```

```python
#17. Cube using lambda
nums = [1, 2, 3, 4]
cubes = apply_func(nums, lambda x: x ** 3)
print(cubes)
```

```python
#18. Remove empty strings
strings = ["hello", "", "world", "", "python"]
result = list(filter(lambda s: s != "", strings))
print(result)
```

```python
#19. List of booleans (positive check)
nums = [-2, 3, -1, 5, 0]
result = list(map(lambda x: x > 0, nums))
print(result)
```

```python
#20. Final Challenge (Mixed strings processing)
data = ["apple", "", "123", "banana", "456", ""]

# Remove empty strings
data = list(filter(lambda x: x != "", data))

# Keep only alphabetic strings
data = list(filter(lambda x: x.isalpha(), data))

# Convert to uppercase
data = list(map(lambda x: x.upper(), data))

print(data)
```
