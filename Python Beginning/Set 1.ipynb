4. Convert Data Types****

```python
num = 50

# int to float
print(float(num))   # 50.0

# int to string
print(str(num))     # "50"

# string to int
s = "123"
print(int(s))       # 123
```

9. Simple Calculator****

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operation (+, -, *, /): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
else:
    print("Invalid operation")
```
