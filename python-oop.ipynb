```python
'''
Q6: Creating Multiple Instances
Problem Statement: Consider a Circle class with a pi class - level attribute, an __init__ method that sets radius, area, and circumference, and methods set_radius and get_circumference. 
Create two Circle objects: circle 1 with radius 10, and circle 2 with radius 2. 
Print their radius and area.
'''

class Circle:
    pi = 3.14

    def __init__(self, radius):
        self.radius = radius
        self.area = self.pi * radius * radius
        self.circumference = 2 * self.pi * radius

    def set_radius(self, radius):
        self.radius = radius

    def get_circumference(self):
        return 2 * self.pi * self.radius

circle1 = Circle(10)
circle2 = Circle(2)

print(circle1.radius, circle1.area)
print(circle2.radius, circle2.area)
```

```python
'''
Q8: Calculating Distance with a Class Method
Problem Statement: Create a class Point for a 2D coordinate system. 
Its __init__ method should take x and y coordinates.
Add a method distance_from_origin to calculate the Euclidean distance from (0,0), using the formula √(x2 −x1)2 +(y2 −y1)2. 
Create a Point at (3, 4) and print its distance.
'''

import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return math.sqrt(self.x**2 + self.y**2)

p = Point(3, 4)
print(p.distance_from_origin())
```

```python
'''
Q10: Calculated Attributes vs. Methods (Line Class)
Problem Statement: Create a class Line that takes two (x,y) coordinate tuples, coord1 and coord2.
Implement distance() to return the Euclidean distance √(x2 −x1)2 +(y2 −y1)2, and slope() to return the slope y2−y1 x2−x1, handling vertical lines. 
Create Line objects for points (1,1) to (4,5) and (0,0) to (0,5), printing distance and slope.
'''

import math

class Line:
    def __init__(self, coord1, coord2):
        self.coord1 = coord1
        self.coord2 = coord2

    def distance(self):
        x1, y1 = self.coord1
        x2, y2 = self.coord2
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    def slope(self):
        x1, y1 = self.coord1
        x2, y2 = self.coord2
        if x2 - x1 == 0:
            return "Undefined"
        return (y2 - y1) / (x2 - x1)

l1 = Line((1,1), (4,5))
l2 = Line((0,0), (0,5))

print(l1.distance(), l1.slope())
print(l2.distance(), l2.slope())
```

```python
'''
Q11: Attribute Update Logic in Methods (Circle Class)
Problem Statement: Modify the set_radius method in the Circle class to update area and circumference when radius changes. 
Create a Circle with radius 5, print initial attributes, change radius to 9, and print updated attributes.
'''

class Circle:
    pi = 3.14

    def __init__(self, radius):
        self.radius = radius
        self.area = self.pi * radius * radius
        self.circumference = 2 * self.pi * radius

    def set_radius(self, radius):
        self.radius = radius
        self.area = self.pi * radius * radius
        self.circumference = 2 * self.pi * radius

c = Circle(5)
print(c.radius, c.area, c.circumference)

c.set_radius(9)
print(c.radius, c.area, c.circumference)
```

```python
'''
Q12: Class and Instance Attributes Interaction
Problem Statement: Create a class Counter with a class-level attribute total_instances initialized to 0.
In __init__, increment total_instances and assign it to an instance attribute id. 
Create three Counter objects and print total_instances and each object’s id.
'''

class Counter:
    total_instances = 0

    def __init__(self):
        Counter.total_instances += 1
        self.id = Counter.total_instances

c1 = Counter()
c2 = Counter()
c3 = Counter()

print(Counter.total_instances)
print(c1.id, c2.id, c3.id)
```

```python
'''
Q13: Polymorphism with Method Overriding
Problem Statement: Define a base class Animal with a make_sound() method that prints “Some generic sound.”. 
Create subclasses Dog and Cat, overriding make_sound() to print “Woof!” and “Meow!”, respectively. 
Create a list of one Animal, one Dog, and one Cat, and call make_sound() on each.
'''

class Animal:
    def make_sound(self):
        print("Some generic sound")

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

animals = [Animal(), Dog(), Cat()]

for a in animals:
    a.make_sound()
```

```python
'''
Q14: Complex Inheritance with super() and Additional Attributes
Problem Statement: Create a base class Product with __init__ taking name and price.
Create a subclass Electronics with __init__ taking name, price, and warranty_months, and a subclass Book taking name, price, and author. 
Use super() in both. Create and print attributes for an Electronics and a Book instance.
'''
    
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Electronics(Product):
    def __init__(self, name, price, warranty_months):
        super().__init__(name, price)
        self.warranty_months = warranty_months

class Book(Product):
    def __init__(self, name, price, author):
        super().__init__(name, price)
        self.author = author

e = Electronics("Laptop", 50000, 24)
b = Book("Python Book", 500, "John Doe")

print(e.name, e.price, e.warranty_months)
print(b.name, b.price, b.author)
```

```python
'''Q15: Flexible Cylinder Class with Dynamic Calculations and Scaling
Problem Statement: Create a Cylinder class with a pi class-level attribute set to 3.14.
Its init should take height and radius. 
Implement calculate_volume() (πr2h) and calculate_surface_area() (2πrh+2πr2).
Add a scale(factor) method to multiply height and radius. 
Create a Cylinder, print initial attributes and calculations, scale by 2, and print updated values.
'''

class Cylinder:
    pi = 3.14

    def __init__(self, height, radius):
        self.height = height
        self.radius = radius

    def calculate_volume(self):
        return self.pi * self.radius**2 * self.height

    def calculate_surface_area(self):
        return 2 * self.pi * self.radius * self.height + 2 * self.pi * self.radius**2

    def scale(self, factor):
        self.height *= factor
        self.radius *= factor

c = Cylinder(5, 2)

print(c.height, c.radius)
print(c.calculate_volume())
print(c.calculate_surface_area())

c.scale(2)

print(c.height, c.radius)
print(c.calculate_volume())
print(c.calculate_surface_area())
```
