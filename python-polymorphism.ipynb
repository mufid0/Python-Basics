```python
#1. Simple Shape Hierarchy & Area Calculation
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


c = Circle(3)
r = Rectangle(4, 5)

print("Circle Area:", c.area())
print("Rectangle Area:", r.area())
```

```python
#2. Basic Bank Account Operations
class BankAccount:
    def __init__(self, initial_balance):
        self.balance = initial_balance

    def deposit(self, amount):
        self.balance += amount
        print("Balance after deposit:", self.balance)

    def withdraw(self, amount):
        self.balance -= amount
        print("Balance after withdrawal:", self.balance)


acc = BankAccount(500)
acc.deposit(200)
acc.withdraw(150)
```

```python
#3. Simple Inventory Item Display
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def get_info(self):
        return f"Product: {self.name}, Price: ${self.price}"


class Service:
    def __init__(self, name, duration_days):
        self.name = name
        self.duration_days = duration_days

    def get_info(self):
        return f"Service: {self.name}, Duration: {self.duration_days} days"


p = Product("Laptop", 1000)
s = Service("Cleaning", 5)

print(p.get_info())
print(s.get_info())
```

```python
#4. Encapsulation (LightBulb Example)

#Encapsulation = hiding internal data and controlling access via methods.

class LightBulb:
    def __init__(self):
        self.__is_on = False  # private variable

    def turn_on(self):
        self.__is_on = True

    def turn_off(self):
        self.__is_on = False

    def status(self):
        return "On" if self.__is_on else "Off"


bulb = LightBulb()
bulb.turn_on()
print(bulb.status())
```

```python
#5. Student Class with __str__
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def __str__(self):
        return f"Student: {self.name}, ID: {self.student_id}"


s = Student("Mufid", 101)
print(s)
```

```python
#6. Default Arguments Function
def display_message(message="Hello there!"):
    print(message)

display_message()
display_message("Good morning!")
```

```python
#7. Operator Overloading (+)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)


p1 = Point(2, 3)
p2 = Point(4, 5)
p3 = p1 + p2

print(p3.x, p3.y)
```

```python
#8. Method Overriding (Animal)
class Animal:
    def make_sound(self):
        print("Generic animal sound.")


class Dog(Animal):
    def make_sound(self):
        print("Woof woof!")


a = Animal()
d = Dog()

a.make_sound()
d.make_sound()
```

```python
#9. Duck Typing
class Square:
    def get_description(self):
        print("I am a square.")


class Circle:
    def get_description(self):
        print("I am a circle.")


def display_shape_info(shape):
    shape.get_description()


display_shape_info(Square())
display_shape_info(Circle())
```

```python
#10. Method Overriding (Instrument)
class Instrument:
    def play(self):
        print("Playing a generic instrument sound.")


class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar!")


class Drums(Instrument):
    def play(self):
        print("Hitting the drums!")


i = Instrument()
g = Guitar()
d = Drums()

i.play()
g.play()
d.play()
```

```python
#11. Function Overloading in Python
'''
Answer:
Python does NOT support traditional function overloading.

Alternative: Default arguments / *args
'''
def add(a, b=0, c=0):
    return a + b + c

print(add(2))
print(add(2, 3))
print(add(2, 3, 4))
```

```python
#13. Polymorphic Area Calculation
class Rectangle:
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w


class Triangle:
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * self.b * self.h


def get_shape_area(shape):
    return shape.area()


r = Rectangle(4, 5)
t = Triangle(3, 6)

print(get_shape_area(r))
print(get_shape_area(t))
```

```python
#14. Power Function
def power(base, exponent=2):
    return base ** exponent

print(power(3))      # square
print(power(2, 3))  # cube
```

```python
#15. Polymorphic Payroll System
class Employee:
    def __init__(self, name):
        self.name = name

    def get_monthly_salary(self):
        return 4000


class HourlyEmployee(Employee):
    def __init__(self, name, hours_worked, hourly_rate):
        super().__init__(name)
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def get_monthly_salary(self):
        return self.hours_worked * self.hourly_rate


class SalariedEmployee(Employee):
    def __init__(self, name, bonus):
        super().__init__(name)
        self.bonus = bonus

    def get_monthly_salary(self):
        return super().get_monthly_salary() + self.bonus


def print_payroll(employees):
    for emp in employees:
        print(emp.name, emp.get_monthly_salary())


employees = [
    Employee("A"),
    HourlyEmployee("B", 100, 10),
    SalariedEmployee("C", 500)
]

print_payroll(employees)
```
